##Performance Metrics - KPIs

### Completion Rate per Step

```python
def calculate_completion_rates(df, steps):

    last_occurrences = {}
    clients_sets = {}

    for step in steps:
        last_occurrences[step] = df[df['process_step'] == step].groupby('client_id').last()
        clients_sets[step] = set(last_occurrences[step].index)

    total_clients_started = len(clients_sets[steps[0]])

    completion_rates = {}

    for i, step in enumerate(steps[1:], start=1):
        total_clients_step = len(clients_sets[step])
        completion_rate_step = total_clients_step / total_clients_started if total_clients_started > 0 else 0
        completion_rates[f" Completion Rate {step}"] = completion_rate_step

    return completion_rates
```

```python
dataframes = {
    'Test': df_test,
    'Control': df_control,
    'No Participation': df_no_participation
}

steps = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

completion_rate_df = pd.DataFrame()

for name, df in dataframes.items():
    completion_rates = calculate_completion_rates(df, steps)
    print(f"Results for {name} group:")
    for step, rate in completion_rates.items():
        print(f"{step}: {rate:.2%}")
    print("\n")

print(completion_rate_df)
```

***Exporting the results as CSV***

---



```python
completion_rate_df = pd.DataFrame()


for name, df in dataframes.items():
    completion_rates = calculate_completion_rates(df, steps)
    completion_rates = pd.DataFrame(completion_rates, index=[name])
    completion_rate_df = pd.concat([completion_rate_df, completion_rates])


completion_rate_df.to_csv('completion_rates.csv', index=True)


print(completion_rate_df)
```

### Time to Conversion

```python
df_test['date_time'] = pd.to_datetime(df_test['date_time'])
df_control['date_time'] = pd.to_datetime(df_control['date_time'])
df_no_participation['date_time'] = pd.to_datetime(df_no_participation['date_time'])
```

```python
def calculate_avg_time_to_conversion(df, max_conversion_time_hours=1):

    # Explicit conversion of 'date_time' column to datetime format
    df['date_time'] = pd.to_datetime(df['date_time'])

    # Filter customers who have completed all steps up to 'confirm'.
    df_conversion = df[df['client_id'].isin(df[df['process_step'] == 'confirm']['client_id'])]

    # Filter data for 'start' and 'confirm' steps
    df_conversion_start = df_conversion[df_conversion['process_step'] == 'start']
    df_conversion_confirm = df_conversion[df_conversion['process_step'] == 'confirm']

    # Merge data to obtain start and confirmation dates for each customer
    df_time_to_conversion = pd.merge(df_conversion_start[['client_id', 'date_time']],
                                     df_conversion_confirm[['client_id', 'date_time']],
                                     on='client_id', suffixes=('_start', '_confirm'))

    # Calculate time to conversion in seconds
    df_time_to_conversion['time_to_conversion'] = (
        df_time_to_conversion['date_time_confirm'] - df_time_to_conversion['date_time_start']
    ).dt.total_seconds()

    # Filter out outliers (e.g. conversion time greater than max_conversion_time_hours)
    df_time_to_conversion = df_time_to_conversion[
        df_time_to_conversion['time_to_conversion'] < max_conversion_time_hours * 3600
    ]

    # Calculate average conversion time in minutes
    avg_time_to_conversion = df_time_to_conversion['time_to_conversion'].mean() / 60  # Conversion in minutes

    return avg_time_to_conversion
```

```python
avg_time_test = calculate_avg_time_to_conversion(df_test)
avg_time_control = calculate_avg_time_to_conversion(df_control)
avg_time_no_participation = calculate_avg_time_to_conversion(df_no_participation)

print(f"Average time to conversion for Test group : {avg_time_test:.2f} minutes")
print(f"Average time to conversion for Control group : {avg_time_control:.2f} minutes")
print(f"Average time to conversion for No Participation group: {avg_time_no_participation:.2f} minutes")
```

### Time Spent on Each Step

```python
import pandas as pd

def time_spent_step(df, threshold=3600):
    df['date_time'] = pd.to_datetime(df['date_time'])
    df = df.sort_values(['client_id', 'date_time'])
    df['previous_time'] = df['date_time'].shift(1)
    df['previous_client'] = df['client_id'].shift(1)
    df['previous_step'] = df['process_step'].shift(1)

    df['time_spent'] = df.apply(
        lambda row: (row['date_time'] - row['previous_time']).total_seconds() if row['client_id'] == row['previous_client'] else None,
        axis=1
    )

    df = df[df['time_spent'] <= threshold]

    df_valid_transitions = df.dropna(subset=['time_spent'])

    avg_time_per_step = df_valid_transitions.groupby('previous_step')['time_spent'].mean()/60

    return avg_time_per_step

```

```python
dataframes = {
    'Group Test': df_test,
    'Group Control': df_control,
    'Group No Participation': df_no_participation
}

for name, df in dataframes.items():
    avg_time = time_spent_step(df)
    print(f"{name} - Time Spent per Step:\n{avg_time}")
```

***Exporting the KPI as CSV***

---



```python

time_spent_df = pd.DataFrame()


for name, df in dataframes.items():
    avg_time = time_spent_step(df)
    avg_time = avg_time.to_frame(name='avg_time_per_step')
    time_spent_df = pd.concat([time_spent_df, avg_time], axis=1)


time_spent_df.to_csv('time_spent_df', index=True)


print(time_spent_df)
```

### Error Rates

```python
#Defining the function that calculates error rate
def detect_backwards_steps(df):

    df['previous_step'] = df['process_step'].shift(1)

    df['previous_client'] = df['client_id'].shift(1)

    df['step_back'] = (df['client_id'] == df['previous_client']) & (df['process_step'] < df['previous_step'])

    backwards_steps = df[df['step_back']]

    clients_backwards = backwards_steps['client_id'].unique()

    clients_started = df[df['process_step'] == 'start']['client_id'].nunique()

    clients_with_errors = len(clients_backwards)

    error_rate = clients_with_errors / clients_started if clients_started > 0 else 0

    return error_rate
```

```python
dataframes = {
    'Group Test': df_test,
    'Group Control': df_control,
    'Group No Participation': df_no_participation
}

for name, df in dataframes.items():
    rate = detect_backwards_steps(df)
    print(f"{name} - Error rate: {rate:.2%}")
```

*Exporting the KPI as CSV*

---

```python
error_rates = {}


for name, df in dataframes.items():
    rate = detect_backwards_steps(df)
    error_rates[name] = rate


error_rate_df = pd.DataFrame(list(error_rates.items()), columns=['Group', 'Error Rate'])
error_rate_df = error_rate_df.set_index('Group')

error_rate_df = error_rate_df.transpose()

error_rate_df.to_csv('/Users/houleyeanne/Documents/GitHub/WEEK5/-vanguard_ab_test_week5_project/tableau/kpis/error_rate_df.csv')

print(error_rate_df)
```

### Error Recovery Rate

```python
def calculate_error_recovery_rate(df):

    # Detect backward steps (previous step > current step)
    df['previous_step'] = df['process_step'].shift(1)
    df['previous_client'] = df['client_id'].shift(1)
    df['step_back'] = (df['client_id'] == df['previous_client']) & (df['process_step'] < df['previous_step'])

    # Clients who made a mistake (backward step)
    clients_with_errors = df[df['step_back']]['client_id'].nunique()

    # Clients who made a mistake and then completed the process
    clients_with_errors_completed = df[(df['client_id'].isin(df[df['step_back']]['client_id'])) &
                                       (df['process_step'] == 'confirm')]['client_id'].nunique()

    # Error recovery rate
    error_recovery_rate = (clients_with_errors_completed / clients_with_errors) if clients_with_errors > 0 else 0

    return error_recovery_rate
```

```python
dataframes = {
    'Group Test': df_test,
    'Group Control': df_control,
    'Group No Participation': df_no_participation
}

for name, df in dataframes.items():
    rate = calculate_error_recovery_rate(df)
    print(f"{name} - Error recovery rate: {rate:.2%}")
```

*Exporting the KPI as CSV*

---

```python
recovery_rates = {}

for name, df in dataframes.items():
    rate = calculate_error_recovery_rate(df)
    recovery_rates[name] = rate

recovery_rate_df = pd.DataFrame(list(recovery_rates.items()), columns=['Group', 'Error Recovery Rate'])
recovery_rate_df = recovery_rate_df.set_index('Group')

recovery_rate_df.to_csv('/Users/houleyeanne/Documents/GitHub/WEEK5/-vanguard_ab_test_week5_project/tableau/kpis/recovery_rate_df.csv')

print(recovery_rate_df)
```

### Drop off rates

```python
def drop_off_rate(df):

    drop_off_rates = {}
    steps = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

    for i in range(len(steps) - 1):
        step_current = steps[i]
        step_next = steps[i + 1]

        clients_current = df[df['process_step'] == step_current]['client_id'].nunique()
        clients_next = df[df['process_step'] == step_next]['client_id'].nunique()

        drop_off_rate = (clients_current - clients_next) / clients_current if clients_current > 0 else 0
        drop_off_rates[step_current] = drop_off_rate

    return drop_off_rates

dataframes = {
    'Group Test': df_test,
    'Group Control': df_control,
    'Group No Participation': df_no_participation
}

print("Drop off rate per step :")

for name, df in dataframes.items():
    rates = drop_off_rate(df)
    print(f"{name}:")

    for step, rate in rates.items():
        print(f"  {step} -> {rate:.2%}")
    print()
```

*Exporting the KPI as CSV*

---

## Hypothesis testing



### Is the difference in completion rate of the the new design and the old design statistically significant ?


- Two-proportions z-test

Hypothesis:
* H‚ÇÄ: There is no difference in completion rates between groups test and control.
* H‚ÇÅ: There is a difference in completion rates between groups test and control.

```python
import numpy as np
import scipy.stats as st
import statsmodels.stats.proportion as proportion


# Number of participants in each group (Test and Control)
n1 = df_test['client_id'].nunique()
n2 = df_control['client_id'].nunique()

# Number of completions in each group
completion_test = df_test[df_test['process_step'] == 'confirm']['client_id'].nunique()
completion_control = df_control[df_control['process_step'] == 'confirm']['client_id'].nunique()

# Count and nobs
count = np.array([completion_test, completion_control])
nobs = np.array([n1, n2])

# Perform proportion test
z_stat, p_value = proportion.proportions_ztest(count, nobs, alternative='two-sided')


print(f"Z-statistic: {z_stat:.2f}, P-value: {p_value:.2f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The difference in completion rates between the Test and Control groups is statistically significant.")
else:
    print("Do not reject the null hypothesis: The difference in completion rates is not statistically significant.")
```

###¬†Checking if the observed increase in completion rate from the A/B test meets or exceeds this 5% threshold.

- Is the completion rate of the Test group significantly higher than the completion rate of the Control group ?

- Two-proportions z-test

Hypothesis:
- H‚ÇÄ: The difference between the completion rate of the Test group and the Control group is less than or equal to 5%.
- H‚ÇÅ: The difference between the completion rate of the Test group and the Control group is greater than 5%.

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

n_test = df_test['client_id'].nunique()
n_control = df_control['client_id'].nunique()


completion_test = df_test[df_test['process_step'] == 'confirm']['client_id'].nunique()
completion_control = df_control[df_control['process_step'] == 'confirm']['client_id'].nunique()

count = np.array([completion_test, completion_control])
nobs = np.array([n_test, n_control])

z_stat, p_value = proportions_ztest(count, nobs, value=0.05, alternative='larger')


print(f"Z-statistic: {z_stat:.2f}, P-value: {p_value:.2f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The completion rate of the Test group is at least 5% higher than the Control group.")
else:
    print("Do not reject the null hypothesis: The completion rate of the Test group is not significantly at least 5% higher.")
```

### Is the average time spent at each step by customers in the Test group is significantly less than that of customers in the Control group ?

- One-sided independent samples t-test

Hypothesis:
- H‚ÇÄ: The average time spent on each step in the Test group is greater or equal to average time in the Control group.
-  H‚ÇÅ: The average time spent on each step in the Test group is less than in the Control group.

```python
from scipy import stats

# Apply the `time_spent_step` function to both Test and Control groups
avg_time_test = time_spent_step(df_test, threshold=3600)
avg_time_control = time_spent_step(df_control, threshold=3600)

# Perform one-sided t-test
t_stat, p_value = stats.ttest_ind(avg_time_test, avg_time_control, alternative='less')

print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: The average time spent per step in the Test group is significantly less than in the Control group.")
else:
    print("Do not reject the null hypothesis: No significant difference in time spent per step between Test and Control groups.")
```

### Is the average age of clients engaging in the new process and the old process is statiscally significant ?


- Test-t de Welch : two-tailed test, we are testing whether the means are different, without specifying a direction.

Hypothesis:
- H‚ÇÄ: Average age of clients engaging in the new process = average age of clients engaging in the old process
- H‚ÇÄ: Average age of clients engaging in the new process != average age of clients engaging in the old process

```python
ages_test = df_test['age']
ages_control = df_control['age']


t_stat, p_value = st.ttest_ind(ages_test, ages_control, equal_var=False)


print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: The difference in mean ages between the Test and Control groups is statistically significant.")
else:
    print("Do not reject the null hypothesis: The difference in mean ages is not statistically significant.")
```

### Is the error rate for clients in the Test group lower than the error rate in Control group ?


- One-sided z-test for proportions

Hypothesis:
- H‚ÇÄ: The error rate for clients in the Test group is less than or equal to the error rate in the Control group.
- H‚ÇÅ: The error rate for clients in the Test group is greater than the error rate in the Control group.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

error_rate_test = detect_backwards_steps(df_test)
error_rate_control = detect_backwards_steps(df_control)

# Number of clients who started in each group
clients_started_test = df_test[df_test['process_step'] == 'start']['client_id'].nunique()
clients_started_control = df_control[df_control['process_step'] == 'start']['client_id'].nunique()

# Number of clients with errors in each group
clients_with_errors_test = df_test[df_test['step_back']]['client_id'].nunique()
clients_with_errors_control = df_control[df_control['step_back']]['client_id'].nunique()

# Counts of errors and number of participants in both groups
count = np.array([clients_with_errors_test, clients_with_errors_control])
nobs = np.array([clients_started_test, clients_started_control])

# Perform a one-sided z-test to check if the Test group's error rate is greater
z_stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"Z-statistic: {z_stat:.2f}, P-value: {p_value:.2f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The Test group has a significantly greater error rate than the Control group.")
else:
    print("Do not reject the null hypothesis: No significant difference in error rates between the Test and Control groups.")
```
